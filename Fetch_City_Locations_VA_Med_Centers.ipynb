```python
s = base_path + "/Veterans Health Administration_ VISN 9_ VA MidSouth Healthcare Network.json"
output_file = join(base_path, "lookup_va_med_city.csv")

if os.path.exists(output_file):
    os.remove(output_file)
```

```python

files_list = []

for f in os.listdir(base_path):
    if isfile(join(base_path, f)):
        if f.endswith(".json"):
            files_list.append(f)
```

```python
all_files = files_list
```

```python
out = pd.DataFrame()

for f in all_files:
    s = base_path + "/" + f
    with open(s) as json_file:
        data = json.load(json_file)
        print("For file %s there are %s number of entries" % (f, len(data)))
        df = pd.DataFrame(data)
        out = out.append(df)
    
    
```
